```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = load_iris()

# Create a Pandas DataFrame from the dataset
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Check for missing values
# No missing values are expected in the Iris dataset, but it's good to double-check
iris_df.isnull().sum()

# Remove duplicate rows
iris_df.drop_duplicates(inplace=True)

# Encode categorical variables (not applicable in this dataset)

# Standardize the numerical features
scaler = StandardScaler()
iris_df[iris.feature_names] = scaler.fit_transform(iris_df[iris.feature_names])

# Explore the cleaned dataset
iris_df.head()
```
